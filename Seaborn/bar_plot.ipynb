```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x='class', y='fare', hue='sex', data=df, order=['Third', 'Second', 'First'], palette='pastel')
plt.show()
```

```python
sns.barplot(x='fare', y='class', hue='sex', data=df)
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, errorbar=None, estimator=np.median, saturation=0.2)
```

```python
sns.barplot(x="class", y="fare", data=df, linewidth=2.5, facecolor=(1, 1, 1, 0), edgecolor=".2")
```
