```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
data['species'].value_counts().sum
```

```python
# FacetGrid
plot = sns.FacetGrid(data=data, col='species')
plot.map(plt.plot, 'sepal_width')
plt.show()
```

```python
# replot method

data = sns.load_dataset('tips')
data.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='time', data=data)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
sns.relplot(x='total_bill', y='tip', hue='day', col='time', row='sex', data=data)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='day', size='size', data=data)
```

```python
markers = {"Lunch": "s", "Dinner": "X"}

sns.scatterplot(x="total_bill", y="tip", style="time", markers=markers, data=data)
plt.show()
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.lineplot(x='timepoint', y='signal', hue='region', style='event', data=fmri)
plt.show()
```
