```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
data['species'].value_counts().sum
```

```python
# FacetGrid
plot = sns.FacetGrid(data=data, col='species')
plot.map(plt.plot, 'sepal_width')
plt.show()
```

```python
# replot method

data = sns.load_dataset('tips')
data.head()
```

```python
data = sns.load_dataset('tips')
graph = sns.FacetGrid(data=data, col='sex', hue='day')
graph.map(plt.scatter, 'total_bill', 'tip')
graph.add_legend()
plt.show()
```

```python
data = sns.load_dataset('tips')
graph = sns.FacetGrid(data=data, row='smoker', col='time')
graph.map(plt.hist, 'total_bill', bins=15)
plt.show()
```

```python
graph = sns.FacetGrid(data=data, col='time', hue='smoker')
graph.map(sns.regplot, 'total_bill', 'tip').add_legend()
plt.show()
```
