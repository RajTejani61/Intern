```python
data = sns.load_dataset('iris')
data.head(2)
```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Iris Dataset')
plt.xlim(5) # x starts from 5
sns.set_style('dark')
```

```python
sns.set_style('darkgrid')
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
plt.show()
```

```python
sns.lineplot(data=data, x='sepal_length', y='sepal_width')
sns.set_style('white')
sns.despine() # remove boundaries of axes 
plt.figure(figsize=(8,12))
plt.show()
```

```python
data = sns.load_dataset('iris')
#sns.set_context('notebook')
def plot():
    sns.lineplot(data=data, x='sepal_length', y='sepal_width')

with sns.axes_style('whitegrid'):
    plt.subplot(2,1,1)
    sns.set_palette('Accent')
    plot()
plt.subplot(2,1,2)
sns.set_palette('vlag')
plot()
plt.show()
```

```python
fig = plt.figure(figsize=(5,4))

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
plot()

ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
plot()

plt.show()
```

```python
ax1 = plt.subplot2grid((6, 1), (0,0), rowspan=2, colspan=1)
plot()
ax2 = plt.subplot2grid((6, 1), (2,0), rowspan=2, colspan=1)
plot()
ax3 = plt.subplot2grid((6, 1), (4,0), rowspan=2, colspan=1)
plot()
```
