```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.catplot(data=data, x='size', y='tip', kind='point')
plt.show()
```

```python
sns.catplot(data=data, x='size', y='tip', hue='sex', kind='point')
```
