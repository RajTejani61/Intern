```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.random.randint(1, 100, (10, 10))
data
```

```python
sns.heatmap(data, annot=True, annot_kws={'size':10})
```

```python
sns.heatmap(data, vmin=20, vmax=80, annot=True)
```

```python
sns.heatmap(data, cmap='coolwarm', annot=True, center=50)
```

```python
import matplotlib.colors as mcolors
sns.heatmap(data, 
            xticklabels=list('ABCDEFGHIJ'), 
            yticklabels=False,
            norm=mcolors.LogNorm(),
            cmap='coolwarm',
            linewidth=0.5,
            annot=True)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Heatmap')
plt.show()
```

```python
data = sns.load_dataset('tips')
data_corr = data.corr(numeric_only=True)
print(data_corr)
sns.heatmap(data_corr, annot=True, cmap='YlGnBu')
```

```python
data = sns.load_dataset('tips')
data = data[['total_bill', 'tip', 'size']]
mask = np.triu(np.ones_like(data.corr()), k=1)
```

```python
sns.heatmap(data.corr(), mask=mask, cmap='YlGnBu', annot=True)
plt.show()
```

```python
data = np.random.rand(12, 12)
sns.heatmap(data, cmap='Greens')
```

```python
# adding frame using axhlin, axvline

graph = sns.heatmap(data, linewidth=0.5)
graph.axhline(y=0, color='k', linewidth=10)
graph.axhline(y=data.shape[1], color='k', linewidth=10)
graph.axvline(x=0, color='k', linewidth=10)
graph.axvline(x=data.shape[0], color='k', linewidth=10)
```

```python
# adding frame using spines

graph = sns.heatmap(data, cmap='coolwarm', linewidth=0.5)

for _, spines in graph.spines.items():
    spines.set_visible(True)
    spines.set_linewidth(5)
    spines.set_linestyle('dashdot')
    
```
