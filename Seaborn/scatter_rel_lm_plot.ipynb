```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(x="total_bill", y="tip", style="time", markers=markers, data=data)
plt.show()
```

```python

df = sns.load_dataset('fmri')
df.head()
```

```python
sns.scatterplot(x='timepoint', y='signal', hue='region', data=df)
plt.title('FMRI signal over time')
plt.show()
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.scatterplot(x='day', y='tip', data=data, marker='+', hue='day', size="size")
plt.title('tips day by day')
plt.show()
```

```python
# rel plot
# Relplot() combines FacetGrid with either of the two axes-level functions scatterplot() and lineplot().
sns.relplot(x='total_bill', y='tip', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='time', data=data)
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
sns.relplot(x='total_bill', y='tip', hue='day', col='time', row='sex', data=data)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', hue='day', size='size', data=data)
```

```python
# lmplot (linear model plot)
# It shows a line representing a linear regression model along with data points on the 2D-space and x and y can be set as the horizontal and vertical labels respectively.
sns.lmplot(x='sepal_length', y='petal_length', data=data, ci=None)
plt.show()
```
