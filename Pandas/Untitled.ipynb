```python
import pandas as pd
import numpy as np
```

```python
#original_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
original_df = pd.read_csv('titanic.csv')
original_df
```

```python
# select useful column
df = original_df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']]
df
```

```python
# filling null age cell

df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
# converting sex into number
df['Sex'] = pd.get_dummies(df['Sex'], drop_first=True, dtype='int')
df.head()
```

```python
df[['Pclass', 'Age', 'Fare']] = (df[['Pclass', 'Age', 'Fare']] - df[['Pclass', 'Age', 'Fare']].mean()) / df[['Pclass', 'Age', 'Fare']].std()
```

```python
X = df[['Pclass', 'Age', 'Fare']].values
Y = df['Survived'].values.reshape(-1, 1)

print(X)
print(Y)
```
