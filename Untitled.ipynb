```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Training data
X = np.array([500, 1000, 1500, 2000]).reshape(-1, 1)
y = np.array([150, 200, 250, 300])  # in $1000s
```

```python
# Create model
model = LinearRegression()
model.fit(X, y)
```

```python
# Predict
y_pred = model.predict(X)
```

```python
print(y_pred)
```

```python
plt.scatter(X, y, color='blue', label='Actual')
plt.plot(X, y_pred, color='red', label='Hypothesis (Line)')
plt.xlabel('Size (sq.ft)')
plt.ylabel('Price ($1000s)')
plt.legend()
plt.title('Linear Regression Hypothesis')
plt.grid(True)
plt.show()
```
