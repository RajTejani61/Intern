Random

```python
from numpy import random as r
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pyt
import random 
```

```python
print(r.randint(100))

print(r.randint(1,101,size=5))
print(r.randint(1,60,10))
print(r.randint(1,50,(2,2)))

#different lower or upper bounds
print(r.randint([1,10,15], 20))
print(r.randint(1,[10,50,100]))
```

```python
r.rand()
```

```python
r.rand(3,3)
```

```python
print(r.choice([1,5,10,15,20], size=5))

arr = np.array(["apple", "banana", "cherry"])
newarr = r.choice(arr, size=2)
print(newarr)
```

```python
x = r.choice([1,2,3,4,5], p=[0.2, 0.3, 0.1, 0.3, 0.1], size=10)
print(x)
```

```python
# Shuffle vs Permutation

#shuffle: It changes in original array
arr = np.array([1,2,3,4,5])
r.shuffle(arr)
print(arr)

#permutation: It doesn't change original array and give new array
new_arr = r.permutation(arr)
print(new_arr)

arr = np.arange(9).reshape(3,3)
print(r.permutation(arr))
```

```python
# normal distribution

arr = r.normal(loc=170, scale=10, size=100)
sns.displot(arr, kind='kde')
pyt.show()
```

```python
# bionomial distribution

arr = r.binomial(10, 0.5,50)
print(arr)
sns.displot(arr, kind='hist')
pyt.show()
```

```python
random.sample(['red', 'blue'], counts=[4, 2], k=5)
```

```python
data = {
    "normal" : r.normal(loc=50, scale=5, size=1000),
    "binomial" : r.binomial(n=100, p=0.5, size=1000)
}
sns.displot(data, kind='kde')
pyt.show()
```

```python
# poission

x = r.poisson(lam=5, size=10)
print(x)
```
