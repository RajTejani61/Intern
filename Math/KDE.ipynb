```python
from sklearn.neighbors import KernelDensity
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
data = np.array([1.33, 0.3, 0.97, 1.1, 0.1, 1.4, 0.4])[:, np.newaxis]

# Fit KDE model
kde = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(data)

# Generate density estimates
x_range = np.linspace(data.min() - 0.3, data.max() + 0.3, 600)[:, np.newaxis]
log_density = kde.score_samples(x_range)

# Plot results
plt.plot(x_range, np.exp(log_density), color='gray', linewidth=2.5)
plt.show()
```
