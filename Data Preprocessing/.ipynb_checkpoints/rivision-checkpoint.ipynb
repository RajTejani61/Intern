```python
import pandas as pd
import scipy
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diabetes.csv')
df.head(3)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(nrows=9, ncols=1, figsize=(7, 17))
i = 0
for col in df.columns:
    ax[i].boxplot(df[col], vert=False)
    ax[i].set_ylabel(col)
    i += 1

plt.show()
```

```python
# clean data
i = 0
for col in df.columns:
    q1, q3 = np.percentile(df[col], [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr*1.5)
    upper_bound = q3 + (iqr*1.5)
    clean_data = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
clean_data.head()
```

```python
corr = clean_data.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr), k=1)
plt.figure(dpi=150)
sns.heatmap(data=corr, mask=mask, annot=True, fmt='.2f')
plt.show()
```

```python
corr['Outcome'].sort_values(ascending=False)
```

```python
plt.pie(clean_data.Outcome.value_counts(), labels=['Diabetes', 'Non Diabetes'], autopct='%.2f', shadow=True)
plt.title('Outcome')
plt.show()
```
