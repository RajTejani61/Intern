```python
df = pd.read_excel('sample_random_data.xlsx')
df
```

```python
# null val
df.isnull().sum()
```

```python
df.dropna()
df.fillna(0)
```

```python
df.rename(columns={'Name' : 'Frist Name', 'Joining Date':'Date of Join'}, inplace=True)
df
```

```python
df['Yearly Income'] = df['Salary']*12
df
```

```python
df['Age Group'] = df['Age'].apply(lambda x: 'Senior' if x > 45 else 'Adult') 
df
```

```python
# pd.cut divide or group the numerical data into categories

df['Marks'] = np.random.randint(0, 100, 10)
bin = [0,20,40,60,80,100] # 0-20, 20-40, 40-60, 60-80, 80-100
label = ['0-20', '20-40', '40-60', '60-80', '80-100']

df['Rank'] = pd.cut(df['Marks'], bins=bin, labels=label)
df
```

```python
# categories Data

mydata = pd.Series(['A', 'B', 'B', 'C', np.nan])
codes, unique = pd.factorize(mydata, use_na_sentinel=True)
print(codes)
print(unique)
```

```python
pd.Categorical(mydata)

mydata.astype('category')
```

```python
# get dummies 

data = {
    'color' : ['red', 'blue', 'green', 'blue', 'red'],
    'size' : ['small', 'large', 'mediam', 'small', 'large']
}
new_df = pd.DataFrame(data)
new_df
```

```python
# one hot encoding
df_encoded = pd.get_dummies(new_df)
df_encoded
```
