```python
import pandas as pd
```

```python
data = {
    'id': [1, 2, 10, 12],
    'val1': ['a', 'b', 'c', 'd']
}

a = pd.DataFrame(data)
a
```

```python
data1 = {
    'id': [1, 2, 9, 8],
    'val1': ['p', 'q', 'r', 's']
}
b = pd.DataFrame(data1)
b
```

```python
df = pd.merge(a, b, on='id', how='inner')
df
```

```python
df = pd.merge(a, b, on='id', how='outer')
df
```

```python
df = pd.merge(a, b, on='id', how='left')
df
```

```python
df = pd.merge(a, b, on='id', how='right')
df
```

```python
df = pd.merge(a, b, left_index=True, right_index=True)
df
```
