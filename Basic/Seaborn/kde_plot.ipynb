```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x= np.random.randn(200)
y = np.random.randn(200)
```

```python
sns.kdeplot(x=x, fill=True)
```

```python
sns.kdeplot(y=y, fill=True, color='green')
```

```python
sns.kdeplot(x, fill=True, color='red')
sns.kdeplot(y, fill=True, color='green')
plt.show()
```

```python
sns.kdeplot(x=x, y=y)
```

```python
sns.kdeplot(x=x, y=y, fill=True, cbar=True)
```

```python
data = sns.load_dataset('iris')
data.species.value_counts()
```

```python
setosa = data.loc[data.species == 'setosa']
virginica = data.loc[data.species == 'virginica']
print(setosa.head())
print(virginica.head())
```

```python
sns.kdeplot(x=setosa.petal_length, y=setosa.petal_width)
sns.kdeplot(x=virginica.sepal_length, y=virginica.sepal_width)
plt.show()
```
