```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('tips')
```

```python
sns.pairplot(data=data, hue='sex')
```

```python
custom_palette = {'Male':'skyblue', 'Female':'lightpink'}
sns.pairplot(data=data, hue='sex', palette=custom_palette)
plt.show()
```

```python
sns.pairplot(data=data, vars=data[['total_bill', 'tip']], diag_kind='kde', markers=['o', 's'])
```

```python
sns.pairplot(data=data, vars=data[['total_bill', 'tip']], hue='size')
```
