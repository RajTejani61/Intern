```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
graph = sns.PairGrid(df, hue ='day')
# graph for diagonal
graph = graph.map_diag(plt.hist)
# graph for non-diagonal
graph = graph.map_offdiag(plt.scatter)
graph = graph.add_legend()
plt.show()
```

```python
 # joint plot (marginal probablity distribution)

data = sns.load_dataset('tips')
data.head()
```

```python
sns.jointplot(data=data, x='total_bill', y='tip', kind='reg', ci=None)
plt.show()
```

```python
sns.jointplot(data=data, x='total_bill', y='tip', hue='time' )
```

```python
# regplot  (regression plot)
data = sns.load_dataset('iris')
data.head()
```

```python
sns.regplot(x='sepal_length', y='petal_length', data=data, ci=None)
plt.show()
```
