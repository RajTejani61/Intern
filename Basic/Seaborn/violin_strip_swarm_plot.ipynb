```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
data.head()
```

```python
sns.violinplot(data=data, x='day', y='tip')
plt.show()
```

```python
sns.violinplot(data=data, x='tip', y='day', linewidth=2)
```

```python
sns.violinplot(x='total_bill', data=data, saturation=0.06)
plt.show()
```

```python
sns.violinplot(data=data, x='day', y='tip', hue='sex')
```

```python
sns.violinplot(data=data, x='day', y='tip', hue='sex', split=True, palette='Set2')
# Instead of creating separate violins for each sub-category, you can split each violin in half using the split=True parameter
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
sns.violinplot(data=data, x='species', y='sepal_length', inner='quart')
```

```python
sns.violinplot(data=data, x='species', y='sepal_length', inner='box')
```

```python
# strip plot

data = sns.load_dataset('tips')
data.head()
```

```python
sns.stripplot(data=data, x='day', y='total_bill', linewidth=2, edgecolor='r')
plt.show()
```

```python
sns.stripplot(data=data, x='day', y='total_bill', jitter=0.2)
plt.show()
```

```python
sns.stripplot(data=data, x='sex', y='total_bill', hue='day', marker='s')
```

```python
# swarm plot

sns.swarmplot(data=data, x='day', y='total_bill')
```

```python
sns.swarmplot(data=data, x='total_bill', y='day', hue='size', palette='Set1')
```

```python
sns.swarmplot(data=data, x='day', y='total_bill', hue='sex', dodge=True)
```

```python
tips = sns.load_dataset("tips")

sns.swarmplot(x="time", y="tip", data=tips, order=["Dinner", "Lunch"], size=5)
plt.show()
```
