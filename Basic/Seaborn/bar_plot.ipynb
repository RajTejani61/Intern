```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
sns.barplot(x='class', y='fare', hue='sex', data=df, order=['Third', 'Second', 'First'], palette='pastel')
plt.show()
```

```python
sns.barplot(x='fare', y='class', hue='sex', data=df)
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, errorbar=None, estimator=np.median, saturation=0.2)
```

```python
sns.barplot(x="class", y="fare", data=df, linewidth=2.5, facecolor=(1, 1, 1, 0), edgecolor=".2")
```

```python
# ascending or descending order

state = ['Rajasthan', 'Bihar', 'Madhya Pradesh', 'Gujarat', 'Maharashtra']
growth = [342239, 94163, 308245, 196024, 307713]

data = pd.DataFrame({'State': state,
                     'Growth': growth})
data
```

```python
sns.barplot(x='State', y='Growth', data=data)
plt.show()
```

```python
data.sort_values('Growth')
```

```python
sns.barplot(x='State', y='Growth', data=data, order=data.sort_values('Growth', ascending=False).State)
```

```python
# count plot

df = sns.load_dataset('tips')

sns.countplot(x='sex', data=df, hue='smoker')
plt.show()
```
