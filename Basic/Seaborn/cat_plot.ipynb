```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("exercise")
data.head()
```

```python
sns.catplot(data=data, x='pulse', y='time', kind='bar')
```

```python
sns.catplot(data=data, 
            x='time', 
            y='pulse', 
            hue='kind')
```

```python
sns.catplot(x="time",
            y="pulse",
            hue="kind",
            data=data, 
            kind="violin")
```

```python
data = sns.load_dataset('titanic')
sns.catplot(data=data, x='sex', hue='survived', kind='count')
```

```python
sns.catplot(x='sex', y='fare', hue='survived', data=data, kind="box")
```

```python
tip = sns.load_dataset('tips')
sns.catplot(x='sex',
            y='tip',
            row='time',
            col='day',
            aspect=0.9,
            dodge=False,
            kind='box',
            data=tip)
```
