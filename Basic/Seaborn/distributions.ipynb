```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# normal distribution
data = np.random.normal(loc=170, scale=10, size=1000)
graph = sns.histplot(data=data, kde=True)
graph.axvline(data.mean(), color='k', linewidth=2, linestyle='dashed')
graph.set_xlabel('Height')
graph.set_ylabel('frequency')
```

```python
# binomial dist.
n = 10
p = 0.5
data = np.random.binomial(n=n, p=p, size=1000)
graph = sns.histplot(data, bins=n+1 ,kde=True, discrete=True)
graph.axvline(data.mean(), color='k', linestyle='dashed')
```

```python
# poisson dist.
no_call = 4 # avg call per hour
size = 1000
data = np.random.poisson(lam=no_call, size=size)
graph = sns.histplot(data, discrete=True)
graph.set_xlabel('No of call per hour')
graph.set_ylabel('frequency')
```

```python
# Values of the random variable (X) and their probabilities
# two coin tossed at same time, no of head appear on coin
x_values = [0, 1, 2] # {TT, (HT, TH), HH}
probabilities = [0.25, 0.5, 0.25]
```

```python
# Plotting the PMF
plt.bar(x_values, probabilities, color='skyblue', alpha=0.7)
plt.xlabel('Number of Heads (X)')
plt.ylabel('Probability P(X=x)')
plt.title('Probability Mass Function (PMF)')
plt.xticks(x_values)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
