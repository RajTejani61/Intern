```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sys
pd.options.mode.copy_on_write = True
```

```python
#original_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df = pd.read_csv('titanic.csv')
df
```

```python
# select useful column
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']]
df
```

```python
df.info(verbose=True)
```

```python
# filling null age cell

df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
# converting sex into number
values =  pd.get_dummies(df['Sex'], drop_first=True, dtype='int')
df['Sex'] = values
df.head()
```

```python
df[['Pclass', 'Age', 'Fare']] = (df[['Pclass', 'Age', 'Fare']] - df[['Pclass', 'Age', 'Fare']].mean()) / df[['Pclass', 'Age', 'Fare']].std()
df
```
